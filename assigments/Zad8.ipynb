```python
from sklearn.cluster import KMeans
kmeans = KMeans(copy_x=True, init='k-means++', max_iter=300, n_clusters=2, n_init=10,
        n_jobs=1, precompute_distances='auto', random_state=None, tol=0.0001,
        verbose=0)
kmeans.fit(college.iloc[:, 2:18])
```

```python
kmeans.cluster_centers_
```

```python
college['Cluster'] = 0
college
```

```python
college['Cluster'] = college['Private'].str.contains('Yes')
college['Cluster'] = college.Cluster.astype(int)
college
```

```python
y_pred = kmeans.predict(college.iloc[:, 2:18])
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(college.Cluster, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(college.Cluster, y_pred))
```
